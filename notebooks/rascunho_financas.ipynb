# 🧪 Laboratório 1: Cadastro de Novas Transações

**Objetivo:** Criar e testar a lógica de captura de dados do usuário para adicionar uma nova despesa à nossa base de dados. Este rascunho será a base da **Opção 1** do nosso Assistente Financeiro.

## 🧠 Conceitos Aplicados neste Bloco
* **Matrizes (Listas de Listas):** Estruturação inicial da nossa base de dados `historico_transacoes`.
* **Captura e Tipagem (`input` e `float`):** Coleta de dados via terminal, garantindo que o valor monetário suporte casas decimais (centavos).
* **Injeção de Dados (`.append()`):** Criação de uma variável temporária `nova_transacao` e inserção na matriz principal.

---

### 📝 Passo a Passo da Lógica:
1. O sistema inicia com duas transações fictícias para facilitar a visualização.
2. O usuário digita a **Descrição**, o **Valor** e a **Categoria**.
3. Esses três dados são agrupados em uma única lista.
4. Essa nova lista é empurrada para o final do nosso histórico.
5. O sistema "printa" a matriz final para provarmos que a inserção funcionou.

````python
# 1. Matriz inicial de teste
historico_transacoes = [
    ["Supermercado", 150.00, "Alimentação"],
    ["Cinema", 45.00, "Lazer"],
]

# === O CÓDIGO COMEÇA AQUI ===
# TODO: Criar um input para a Descrição (string)
descricao = input("Descrição do gasto: ")
# TODO: Criar um input para o Valor (convertido para float!)
valor = float(input("Valor gasto: "))
# TODO: Criar um input para a Categoria (string)
categoria = input("Categoria do gasto: ")

# TODO: Criar uma lista chamada 'nova_transacao' com as 3 variáveis acima
nova_transacao = [descricao, valor, categoria]
# TODO: Usar o .append() para jogar a 'nova_transacao' dentro do 'historico_transacoes'
historico_transacoes.append(nova_transacao)

# 2. Linha para testar se funcionou
print(historico_transacoes)
````

# 🧪 Laboratório 2: Extrato Detalhado

**Objetivo:** Desenvolver e validar a lógica da **Opção 2** do nosso sistema, transformando a visualização bruta da matriz em um extrato de fácil leitura para o usuário final.

## 🧠 Conceitos Aplicados neste Bloco
* **Estruturas de Repetição (`for`):** Iteração sobre cada sublista dentro da matriz principal `historico_transacoes`.
* **Indexação:** Acesso aos dados específicos de cada transação utilizando seus índices (`[0]` para descrição, `[1]` para valor, `[2]` para categoria).
* **Formatação de Strings (`f-strings`):** Formatação profissional de saída, incluindo o símbolo de moeda (R$) e separadores.

---

### 📝 Passo a Passo da Lógica:
1. O sistema acessa a matriz `historico_transacoes` (que contém os dados acumulados).
2. O loop percorre cada transação individualmente.
3. Para cada transação, os elementos são extraídos e formatados em uma única linha de texto.
4. O resultado é exibido no terminal de forma organizada.

```python
# === Laboratório 2: Extrato Completo ===
print("\n--- EXTRATO DE TRANSAÇÕES ---")

# Exemplo de loop para percorrer a matriz e formatar a saída
for transacao in historico_transacoes:
    # Acesso aos dados via índices: 0=descrição, 1=valor, 2=categoria
    print(f"{transacao[0]} - R$ {transacao[1]:.2f} (Categoria: {transacao[2]})")

````python
# === Laboratório 2: Extrato Completo ===
print("\n--- EXTRATO DE TRANSAÇÕES ---")

# === O CÓDIGO COMEÇA AQUI ===
for transacao in historico_transacoes:
        print("Descrição: {} - Valor gasto: {:.2f} - Categoria: {}" .format(transacao[0], transacao[1], transacao[2]))
````
